```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('E:\\ArticleClassifier\\Resources\\author_article_gender.csv')
```

```python
df.set_index('Unnamed: 0')
df=df.dropna()
df.head()
```

```python
from sklearn.svm import LinearSVC # Linear support vector classification
```

### Preprocessing & Splitting of the dataset 

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
x , y  = df['article'],df['gender']
y
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

### Count Vectorizer of fitting and transforming 

```python
countVectorizer = CountVectorizer(stop_words='english',max_df=0.8,lowercase = False)
x = countVectorizer.fit_transform(x)
```

```python
countVectorizer.vocabulary_ # To check the vocablory stuff
```

```python
train_x, test_x,train_y,test_y  = train_test_split(x,y)

```

```python
lr = LogisticRegression(solver='newton-cg')
```

```python
lr.fit(train_x,train_y)
```

```python
lr.classes_ # Just checking the classes in there
```

```python
lr.score(test_x,test_y)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y,lr.predict(test_x),labels = lr.classes_)
print(lr.classes_)
cm
#cm
```

```python
lr = LogisticRegression(solver='newton-cg')
cv_score = cross_val_score(lr,X=x,y=y,cv = 10)
print(np.min(cv_score),np.max(cv_score),np.mean(cv_score))
```
